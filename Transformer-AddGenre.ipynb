# Data Preprocessing

```python
import pandas as pd
from datetime import datetime
def normalize_date(date_str):
    try:
        return pd.to_datetime(date_str, format='%b %d, %Y')
    except ValueError:
        return pd.to_datetime(f'{date_str}, 2025', format='%b %d, %Y')
train['Date'] = train['Date'].apply(normalize_date)
```

```python
# Fungsi untuk mengubah Views ke angka
def clean_views(view_str):
    view_str = view_str.lower().replace(" views", "").strip()
    if 'k' in view_str:
        return float(view_str.replace('k', '')) * 1000
    elif 'm' in view_str:
        return float(view_str.replace('m', '')) * 1_000_000
    else:
        return float(view_str.replace(',', ''))
train['Views'] = train['Views'].apply(clean_views)
```

```python
import pandas as pd

all_words = ' '.join(train['Title'].astype(str)).lower().split()
import re
all_words_clean = [re.sub(r'\W+', '', word) for word in all_words if re.sub(r'\W+', '', word) != '']

unique_words = set(all_words_clean)

print(f"Jumlah kata unik di kolom Title: {len(unique_words)}")
```

```python
from collections import Counter

word_counts = Counter(all_words_clean)
most_common = word_counts.most_common(100)
print("20 kata yang paling sering muncul:")
for word, count in most_common:
    print(f"{word}: {count}")
```

```python
least_common = word_counts.most_common()[::-1]  # Balik urutan
print("20 kata yang paling jarang muncul:")
for word, count in least_common[:20]:
    print(f"{word}: {count}")
```

```python
import re
import pandas as pd

# Kata-kata yang ingin dihapus
stop_words_custom = {
    "1", "2", "3", "5", "7", "a", "soal", "if2140", "04float"
}

# Fungsi untuk membersihkan kata-kata dalam Title
def clean_title(title):
    if pd.isna(title):
        return []
    
    # Ubah ke huruf kecil
    title = title.lower()
    
    # Hapus angka desimal seperti 10.15, 8.2, dll
    title = re.sub(r'\b\d+\.\d+\b', '', title)
    
    # Hapus angka bulat seperti 01, 9, 100
    title = re.sub(r'\b\d+\b', '', title)
    
    # Ambil hanya kata dan huruf/angka
    words = re.findall(r'\b[a-zA-Z0-9]+\b', title)
    
    # Hapus stopword dan angka murni
    return [word for word in words if word not in stop_words_custom and not word.isdigit()]

# Terapkan ke kolom Title dan simpan hasil bersihannya
train["Title"] = train["Title"].apply(clean_title)
train["Title"] = train["Title"].apply(lambda x: " ".join(x))
```

# Model

```python
from bertopic import BERTopic
from bertopic import BERTopic
from umap import UMAP
from hdbscan import HDBSCAN

# Set komponen UMAP dan HDBSCAN dengan random_state
umap_model = UMAP(n_neighbors=15, n_components=5, min_dist=0.0, metric='cosine', random_state=42)
hdbscan_model = HDBSCAN(min_cluster_size=3, prediction_data=True)

model = BERTopic(
    language="multilingual",
    nr_topics=13,
    embedding_model="paraphrase-multilingual-MiniLM-L12-v2",
    umap_model=umap_model,
    hdbscan_model=hdbscan_model,
    calculate_probabilities=True,
    verbose=True
)

topics, probs = model.fit_transform(titles)


# Lihat hasil prediksi topik (genre) untuk tiap judul
train['Predicted_Topic'] = topics
train['Topic_Name'] = train['Predicted_Topic'].apply(lambda x: model.get_topic(x))
print(train[['Title', 'Predicted_Topic', 'Topic_Name']])
```
